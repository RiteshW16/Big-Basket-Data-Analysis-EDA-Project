```python
#filing the nulll values with 0

df['product'].fillna(0,inplace=True)
df['brand'].fillna(0,inplace=True)
df['rating'].fillna(0,inplace=True)
df['description'].fillna(0,inplace=True)
```

```python
a=df.groupby('brand')['sale_price'].sum().reset_index()
a=a.sort_values(by='sale_price',ascending=False)
a=a.head(10)
a
```

```python
plt.figure(figsize=(9,9))
sns.barplot(x=a['brand'],y=a['sale_price'])
plt.xticks( rotation=45)
plt.xlabel('Brand')
plt.ylabel('Sale_price')
plt.title('Top 10 Brands with Heighest Price',)
plt.show()
```

Prestige is the most expensive brand, followed by BB Royal and Dkny.The prices of the brands on the list range from around 20,000 to 140,000.

```python
b=df.groupby('category')['sale_price'].sum().reset_index()
b=b.sort_values(by='sale_price',ascending=True)
b=b.head(5)
b
```

```python
c=df.groupby('sub_category')['sale_price'].sum().reset_index()
c=c.sort_values(by='sale_price',ascending=True)
c=c.head(5)
c
```

```python
plt.figure(figsize=(6,6))
exp=(0.04,0,0,0,0)
colors=['#ff3333','#ffad33','#3385ff','#339966','violet']
plt.title("Top 5 Product Sub-Category Wise Profits",fontsize=15)
plt.pie(c['sale_price'],labels=c['sub_category'],autopct="%1.2f%%",colors=colors,explode=exp)
plt.show()
```

In this chart, the top 5 product subcategories contribute the most profits, with organic Fruits & Vegetables and pork & other Meats contributing more profits compared to others.
